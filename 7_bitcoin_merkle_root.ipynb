```python
import hashlib

block = {
    "version": 1,
    "previous_block_hash": "0000000000000000000abc123",
    "timestamp": 1712000000,
    "difficulty_target": "1d00ffff",
    "nonce": 2083236893
}

transactions = [
    "Alice->Bob:2 BTC",
    "Bob->Charlie:0.5 BTC",
    "Charlie->Dave:0.2 BTC",
    "Miner->Alice:6.25 BTC"
]
```

```python
def double_sha256(data):
    return hashlib.sha256(hashlib.sha256(data).digest()).hexdigest()
```

```python
def merkle_root(tx_list):
    current = [double_sha256(tx.encode()) for tx in tx_list]

    while len(current) > 1:
        if len(current) % 2 == 1:
            current.append(current[-1])

        next_level = []
        for i in range(0, len(current), 2):
            pair = (current[i] + current[i + 1]).encode()
            next_level.append(double_sha256(pair))

        current = next_level

    return current[0]
```

```python
print("Bitcoin Block Structure:")
for k, v in block.items():
    print(k, ":", v)

```

```python
print("\nTransactions:")
for i, tx in enumerate(transactions, start=1):
    print(i, tx)
```

```python

root = merkle_root(transactions)
print("\nComputed Merkle Root:", root)
```
