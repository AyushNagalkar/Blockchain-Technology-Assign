# Assignment 5: Deploy HelloWorld Smart Contract on Ethereum Testnet

## Method: Using Remix IDE (Simplest Approach)

---

## Step 1: Setup MetaMask Wallet

1. Install **MetaMask** browser extension from https://metamask.io
2. Create a new wallet and save your seed phrase
3. Switch network to **Sepolia Testnet**:
   - Click network dropdown (top of MetaMask)
   - Select "Sepolia test network"

## Step 2: Get Free Test ETH

Get free Sepolia ETH from a faucet:
- https://sepoliafaucet.com
- https://www.alchemy.com/faucets/ethereum-sepolia
- https://cloud.google.com/application/web3/faucet/ethereum/sepolia

## Step 3: Open Remix IDE

1. Go to https://remix.ethereum.org
2. Create new file: `HelloWorld.sol`
3. Paste the contract code below:

````python
# HelloWorld.sol - Copy this to Remix IDE

contract_code = '''
// SPDX-License-Identifier: MIT
pragma solidity ^0.8.0;

contract HelloWorld {
    string public message;

    // Constructor sets initial message
    constructor() {
        message = "Hello, World!";
    }

    // Function to get the message
    function getMessage() public view returns (string memory) {
        return message;
    }

    // Function to set a new message
    function setMessage(string memory newMessage) public {
        message = newMessage;
    }
}
'''

print(contract_code)
````

## Step 4: Compile the Contract

1. In Remix, click **Solidity Compiler** (left sidebar, 2nd icon)
2. Select compiler version: `0.8.0` or higher
3. Click **Compile HelloWorld.sol**
4. Green checkmark = success!

## Step 5: Deploy to Sepolia Testnet

1. Click **Deploy & Run** (left sidebar, 3rd icon)
2. Change **Environment** dropdown to: `Injected Provider - MetaMask`
3. MetaMask popup will ask to connect - click **Connect**
4. Ensure it shows "Sepolia" network
5. Click **Deploy** button
6. Confirm transaction in MetaMask popup
7. Wait for confirmation (~15-30 seconds)

## Step 6: Interact with Deployed Contract

After deployment, you'll see your contract under "Deployed Contracts":

1. Click **getMessage** - Returns "Hello, World!"
2. Enter new text in **setMessage** field, click button
3. Confirm transaction in MetaMask
4. Click **getMessage** again - Shows your new message!

## Step 7: Verify on Etherscan

1. Copy the contract address from Remix
2. Go to https://sepolia.etherscan.io
3. Paste your contract address in search
4. View your deployed contract and transactions!
