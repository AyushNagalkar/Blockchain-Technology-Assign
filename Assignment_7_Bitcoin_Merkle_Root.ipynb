# 📘 Assignment 7 - Analyze Bitcoin Block Structure and Compute Merkle Root for Sample Transactions

## 🎯 Objective
To understand the structure of a Bitcoin block and implement the computation of the Merkle Root from a list of transactions using hashing.

## 🧠 Concepts Used
1. Bitcoin Block Structure
   A block in Bitcoin contains block header, previous block hash, timestamp, nonce, Merkle root, and transaction list.
2. Hash Function
   Converts input to fixed-size output using SHA-256.
3. Merkle Tree
   Binary tree of transaction hashes used for efficient verification and data integrity.
4. Merkle Root
   Final hash at the top of the tree representing all transactions.

## ⚙️ Flow of Process
Take List of Transactions
        ↓
Hash Each Transaction (SHA-256)
        ↓
Pair Adjacent Hashes
        ↓
Hash Each Pair Together
        ↓
Repeat Until One Hash Remains
        ↓
Final Hash = Merkle Root

## 🔹 Step 1: Import Library

```python
import hashlib
```

## 🔹 Step 2: Hash Function

```python
def sha256(data):
    return hashlib.sha256(data.encode()).hexdigest()
```

## 🔹 Step 3: Compute Merkle Root

```python
def compute_merkle_root(transactions):
    hashes = [sha256(tx) for tx in transactions]

    print("Initial Transaction Hashes:")
    for h in hashes:
        print(h)

    while len(hashes) > 1:
        temp = []

        for i in range(0, len(hashes), 2):
            if i + 1 < len(hashes):
                combined = hashes[i] + hashes[i + 1]
            else:
                combined = hashes[i] + hashes[i]

            new_hash = sha256(combined)
            temp.append(new_hash)

        hashes = temp

        print("\nNew Level Hashes:")
        for h in hashes:
            print(h)

    return hashes[0]
```

## 🔹 Step 4: Sample Transactions

```python
transactions = [
    "Alice pays Bob 10 BTC",
    "Bob pays Charlie 5 BTC",
    "Charlie pays Dave 2 BTC",
    "Dave pays Eve 1 BTC"
]
```

## 🔹 Step 5: Compute Merkle Root

```python
merkle_root = compute_merkle_root(transactions)

print("\nMerkle Root:", merkle_root)
```
