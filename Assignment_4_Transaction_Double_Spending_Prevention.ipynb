# Assignment 4 - Transaction + Double Spending Prevention

## Concept (Short Theory)
A wallet has:
- Public key: acts like address
- Private key: used to sign transactions

A transaction includes:
- Sender
- Receiver
- Amount
- Signature

Double spending problem:
- Same money used twice
- Prevented by checking balance before transaction

## 1. Import Libraries

```python
import hashlib
import json
```

## 2. Wallet Class (Simple Simulation)

```python
class Wallet:
    def __init__(self, name):
        self.name = name
        self.balance = 100  # initial balance

    def __str__(self):
        return f"{self.name} | Balance: {self.balance}"
```

## 3. Transaction Class

```python
class Transaction:
    def __init__(self, sender, receiver, amount):
        self.sender = sender
        self.receiver = receiver
        self.amount = amount
        self.signature = self.sign_transaction()

    def sign_transaction(self):
        # simple hash as signature
        data = self.sender.name + self.receiver.name + str(self.amount)
        return hashlib.sha256(data.encode()).hexdigest()

    def __str__(self):
        return f"{self.sender.name} -> {self.receiver.name} : {self.amount}"
```

## 4. Blockchain (Transaction Validation)

```python
class SimpleBlockchain:
    def __init__(self):
        self.transactions = []

    def add_transaction(self, transaction):
        # Check balance before adding transaction
        if transaction.sender.balance < transaction.amount:
            print("Transaction Failed: Insufficient Balance (Double Spending Prevention)")
            return False

        # Process transaction
        transaction.sender.balance -= transaction.amount
        transaction.receiver.balance += transaction.amount

        self.transactions.append(transaction)
        print("Transaction Successful")
        return True
```

## 5. Create Wallets

```python
alice = Wallet("Alice")
bob = Wallet("Bob")

print(alice)
print(bob)
```

## 6. Initialize Blockchain

```python
blockchain = SimpleBlockchain()
```

## 7. Valid Transaction

```python
tx1 = Transaction(alice, bob, 50)
blockchain.add_transaction(tx1)

print(alice)
print(bob)
```

## 8. Attempt Double Spending

```python
# Alice now has only 50 left, trying to send 70
tx2 = Transaction(alice, bob, 70)
blockchain.add_transaction(tx2)

print(alice)
print(bob)
```
