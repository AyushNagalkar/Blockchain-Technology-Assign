# 📘 Assignment 9 - Deploying Public Key Infrastructure (PKI) Based Identity System

## 🎯 Objective
To implement a Public Key Infrastructure (PKI) based identity system for secure authentication using public-private key cryptography.

## 🧠 Concepts Used
1. Public Key Infrastructure (PKI): Framework for secure digital identity management using public key, private key, and certificates.
2. Public Key Cryptography: Public key is shared openly, private key is kept secret.
3. Digital Signature: Used to verify identity and ensure authentication and integrity.
4. Certificate Authority (CA): Trusted entity that verifies identities and issues digital certificates.
5. Hashing (SHA-256): Used before signing data to ensure fixed-size secure input.

## ⚙️ Flow of Process
Generate Public & Private Keys
        ↓
User Registers Identity (Public Key Stored)
        ↓
User Signs Message Using Private Key
        ↓
System Verifies Signature Using Public Key
        ↓
If Valid → Identity Verified
        ↓
If Invalid → Access Denied

## 🔹 Step 2: Import Libraries

```python
from cryptography.hazmat.primitives.asymmetric import rsa, padding
from cryptography.hazmat.primitives import hashes, serialization
```

## 🔹 Step 3: Generate Keys (User Identity)

```python
def generate_keys():
    private_key = rsa.generate_private_key(
        public_exponent=65537,
        key_size=2048
    )

    public_key = private_key.public_key()

    return private_key, public_key
```

## 🔹 Step 4: Sign Data (Authentication)

```python
def sign_message(private_key, message):
    signature = private_key.sign(
        message.encode(),
        padding.PSS(
            mgf=padding.MGF1(hashes.SHA256()),
            salt_length=padding.PSS.MAX_LENGTH
        ),
        hashes.SHA256()
    )
    return signature
```

## 🔹 Step 5: Verify Signature

```python
def verify_signature(public_key, message, signature):
    try:
        public_key.verify(
            signature,
            message.encode(),
            padding.PSS(
                mgf=padding.MGF1(hashes.SHA256()),
                salt_length=padding.PSS.MAX_LENGTH
            ),
            hashes.SHA256()
        )
        return True
    except:
        return False
```

## 🔹 Step 6: Simulate PKI Identity System

```python
# Generate identity
private_key, public_key = generate_keys()

message = "User Authentication Request"

# Sign message
signature = sign_message(private_key, message)

# Verify identity
is_valid = verify_signature(public_key, message, signature)

print("Signature Valid:", is_valid)
```

## 🔹 Step 7: Test Invalid Case ❌

```python
fake_message = "Hacker Message"

is_valid = verify_signature(public_key, fake_message, signature)

print("Signature Valid (Fake):", is_valid)
```
