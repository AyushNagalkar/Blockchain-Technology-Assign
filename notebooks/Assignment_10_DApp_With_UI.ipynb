## 📄 Step 2: Full UI App Code

````python
import streamlit as st
from web3 import Web3

# -----------------------------
# 🔗 Connect to Ethereum
# -----------------------------
infura_url = "https://sepolia.infura.io/v3/YOUR_PROJECT_ID"
w3 = Web3(Web3.HTTPProvider(infura_url))

# -----------------------------
# 📄 Contract Details
# -----------------------------
contract_address = "YOUR_CONTRACT_ADDRESS"

abi = [
    {
        "inputs": [],
        "name": "getMessage",
        "outputs": [{"type": "string"}],
        "stateMutability": "view",
        "type": "function"
    },
    {
        "inputs": [{"name": "newMessage", "type": "string"}],
        "name": "setMessage",
        "outputs": [],
        "stateMutability": "nonpayable",
        "type": "function"
    }
]

contract = w3.eth.contract(address=contract_address, abi=abi)

# -----------------------------
# 🔐 Wallet Setup
# -----------------------------
account = "YOUR_WALLET_ADDRESS"
private_key = "YOUR_PRIVATE_KEY"

# -----------------------------
# 🎨 UI Design
# -----------------------------
st.set_page_config(page_title="Blockchain DApp", layout="centered")

st.markdown("""
<style>
    .main {
        background: linear-gradient(135deg, #0f172a, #111827, #1f2937);
        color: white;
    }
    .stButton>button {
        border-radius: 12px;
        border: none;
        padding: 0.6rem 1rem;
        font-weight: 700;
        background: linear-gradient(90deg, #22c55e, #06b6d4);
        color: white;
    }
    .block-container {
        padding-top: 2rem;
        padding-bottom: 2rem;
    }
</style>
""", unsafe_allow_html=True)

st.title("🚀 Simple Blockchain DApp")
st.subheader("Interact with a Smart Contract in Style")
st.caption("Built with Streamlit + Web3.py + Sepolia Test Network")

# -----------------------------
# 📥 Read Data
# -----------------------------
if st.button("Get Current Message"):
    message = contract.functions.getMessage().call()
    st.success(f"📢 Message: {message}")

# -----------------------------
# ✍️ Input Field
# -----------------------------
new_message = st.text_input("Enter New Message")

# -----------------------------
# 📤 Send Transaction
# -----------------------------
if st.button("Update Message"):
    if new_message == "":
        st.warning("⚠️ Please enter a message")
    else:
        nonce = w3.eth.get_transaction_count(account)

        txn = contract.functions.setMessage(new_message).build_transaction({
            'from': account,
            'nonce': nonce,
            'gas': 200000,
            'gasPrice': w3.to_wei('10', 'gwei')
        })

        signed_txn = w3.eth.account.sign_transaction(txn, private_key)
        tx_hash = w3.eth.send_raw_transaction(signed_txn.rawTransaction)

        st.success(f"✅ Transaction Sent: {tx_hash.hex()}")


# -----------------------------
# 📌 Footer
# -----------------------------
st.info("Use testnet only. Never expose private keys publicly.")
````
