### 🔹 Step 2: Sample Smart Contract Data

````python
contracts = [
    {
        "name": "HelloWorld",
        "address": "0x123",
        "functions": ["getMessage", "setMessage"]
    },
    {
        "name": "SimpleToken",
        "address": "0x456",
        "functions": ["transfer", "getBalance"]
    },
    {
        "name": "VotingSystem",
        "address": "0x789",
        "functions": ["vote", "getResults"]
    }
]
````

### 🔹 Step 3: Search Function

````python
def search_contracts(keyword):
    results = []

    for contract in contracts:
        if keyword.lower() in contract["name"].lower():
            results.append(contract)
        else:
            for func in contract["functions"]:
                if keyword.lower() in func.lower():
                    results.append(contract)
                    break

    return results
````

### 🔹 Step 4: Build UI (Streamlit App)

````python
import streamlit as st

st.set_page_config(page_title="Smart Contract Search Engine")

st.title("🔍 Smart Contract Search Engine")
st.subheader("Search contracts by name or function")

keyword = st.text_input("Enter keyword (function / contract name)")

if st.button("Search"):
    results = search_contracts(keyword)

    if results:
        st.success(f"Found {len(results)} result(s)")

        for contract in results:
            st.markdown(f"### 📄 {contract['name']}")
            st.write("📍 Address:", contract["address"])
            st.write("⚙️ Functions:", ", 
functions"]))
            st.write("---")
    else:
        st.error("No matching contracts found")
markdown
markdown
### 🔹 Step 5: Run the App

```bash
streamlit run search_app.py
```

## 🖥️ UI Features
- Search bar
- Result cards
- Function-based filtering
- Clean interface

## ✅ Sample Output
Search: transfer

Result:
📄 SimpleToken
📍 Address: 0x456
⚙️ Functions: transfer, getBalance

## 🧾 Result
Smart contracts indexed successfully. Search functionality implemented.

## ⚠️ Limitations
Uses a static dataset in this version. Can be improved with Etherscan API or database integration.

## 🔍 Conclusion
This project demonstrates efficient blockchain data searching with a UI and is a strong interview-level project.
````
