```python
primes = []
pairs = []


# Function to check prime
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
```

```python

# Function to find prime pairs
def prime_pairs(x, r):
    
    # Step 1: generate primes up to r
    for i in range(2, r + 1):
        if is_prime(i):
            primes.append(i)

    # Step 2: check prime pairs
    
    for i in range(len(primes)):
        
        for j in range(i, len(primes)):   # avoids duplicate pairs
            
            if primes[i] * primes[j] > x:
                pairs.append((primes[i], primes[j]))
                
    return 
```
