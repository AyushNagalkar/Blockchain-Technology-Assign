```python
import hashlib
```

```python
def sha256(data):
    return hashlib.sha256(data.encode()).hexdigest()
```

```python
class Block:
    
    def __init__(self, index, prev , data , nonce=0):
        self.index = index
        self.prev = prev
        self.data = data
        self.nonce = nonce
        self.hash = self.compute_hash()
        
    def compute_hash(self):
        print("Computing hash...")
        string = f"{self.index}{self.prev}{self.data}{self.nonce}"
        print(string)
        return sha256(string)
```

```python
b = Block(0, "0", "First Block")
computed_hash = b.compute_hash()
print("Block Hash:", computed_hash)
```
