# 📘 Assignment 8 - Implement a Simple Proof-of-Work (PoW) Mechanism for Mining Blocks

## 🎯 Objective
To implement a Proof-of-Work (PoW) algorithm used in Bitcoin mining to validate blocks and ensure security in a blockchain.

## 🧠 Concepts Used
1. Proof-of-Work (PoW): A consensus mechanism requiring solving a computational puzzle.
2. Mining: Process of finding a valid hash by changing nonce repeatedly.
3. Nonce: A number changed repeatedly to generate different hashes.
4. Difficulty Level: Defines how hard mining is, such as requiring the hash to start with '0000'.
5. SHA-256 Hashing: Used to generate block hashes.

## ⚙️ Flow of Process
Create Block Data
        ↓
Set Nonce = 0
        ↓
Generate Hash (Block Data + Nonce)
        ↓
Check Difficulty Condition
        ↓
If NO → Increment Nonce and Repeat
        ↓
If YES → Block Mined Successfully

## 🔹 Step 1: Import Library

```python
import hashlib
```

## 🔹 Step 2: Block Class

```python
class Block:
    def __init__(self, data, previous_hash):
        self.data = data
        self.previous_hash = previous_hash
        self.nonce = 0
        self.hash = self.calculate_hash()

    def calculate_hash(self):
        value = str(self.data) + str(self.previous_hash) + str(self.nonce)
        return hashlib.sha256(value.encode()).hexdigest()
```

## 🔹 Step 3: Proof-of-Work Function

```python
def mine_block(block, difficulty):
    target = "0" * difficulty

    print("Mining started...")

    while not block.hash.startswith(target):
        block.nonce += 1
        block.hash = block.calculate_hash()

    print("Block mined!")
    print("Nonce:", block.nonce)
    print("Hash:", block.hash)
```

## 🔹 Step 4: Create and Mine Block

```python
difficulty = 4  # number of leading zeros required

block = Block("Alice pays Bob 10 BTC", "0000abc123")

mine_block(block, difficulty)
```
