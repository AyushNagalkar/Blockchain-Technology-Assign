```python
import hashlib
import time

class Block:
    def __init__(self, index, previous_hash, data):
        self.index = index
        self.previous_hash = previous_hash
        self.data = data
        self.timestamp = time.time()
        self.nonce = 0
        self.hash = self.calculate_hash()

    def calculate_hash(self):
        text = f"{self.index}{self.previous_hash}{self.data}{self.timestamp}{self.nonce}"
        return hashlib.sha256(text.encode()).hexdigest()
```

```python
def mine_block(block, difficulty):
    target = "0" * difficulty

    while not block.hash.startswith(target):
        block.nonce += 1
        block.hash = block.calculate_hash()

    return block
```

```python
difficulty = 4
```

```python

genesis = Block(0, "0", "Genesis Block")
block1 = Block(1, genesis.hash, "Alice pays Bob 1 BTC")
```

```python

mined = mine_block(block1, difficulty)
```

```python

print("Mined Block Details:")
print("Index:", mined.index)
print("Previous Hash:", mined.previous_hash)
print("Data:", mined.data)
print("Nonce:", mined.nonce)
print("Hash:", mined.hash)
print("Valid PoW:", mined.hash.startswith("0" * difficulty))
```
