## 🔹 Step 2: Import Libraries

```python
from web3 import Web3
from solcx import compile_source, install_solc
```

## 🔹 Step 3: Install Solidity Compiler

```python
install_solc('0.8.0')
```

## 🔹 Step 4: Connect to Ethereum Network

```python
infura_url = "https://sepolia.infura.io/v3/YOUR_PROJECT_ID"
w3 = Web3(Web3.HTTPProvider(infura_url))

print("Connected:", w3.is_connected())
```

## 🔹 Step 5: Wallet Configuration

```python
account = "YOUR_WALLET_ADDRESS"
private_key = "YOUR_PRIVATE_KEY"
```

## 🔹 Step 6: Write Token Smart Contract

```python
contract_source_code = '''
pragma solidity ^0.8.0;

contract SimpleToken {
    mapping(address => uint) public balances;
    uint public totalSupply;

    constructor(uint initialSupply) {
        balances[msg.sender] = initialSupply;
        totalSupply = initialSupply;
    }

    function transfer(address receiver, uint amount) public {
        require(balances[msg.sender] >= amount, "Insufficient Balance");

        balances[msg.sender] -= amount;
        balances[receiver] += amount;
    }

    function getBalance(address account) public view returns (uint) {
        return balances[account];
    }
}
'''
```

## 🔹 Step 7: Compile Contract

```python
compiled_sol = compile_source(contract_source_code)
contract_id, contract_interface = compiled_sol.popitem()

bytecode = contract_interface['bin']
abi = contract_interface['abi']
```

## 🔹 Step 8: Deploy Contract

```python
Token = w3.eth.contract(abi=abi, bytecode=bytecode)

nonce = w3.eth.get_transaction_count(account)

transaction = Token.constructor(1000).build_transaction({
    'from': account,
    'nonce': nonce,
    'gas': 2000000,
    'gasPrice': w3.to_wei('10', 'gwei')
})

signed_txn = w3.eth.account.sign_transaction(transaction, private_key)
tx_hash = w3.eth.send_raw_transaction(signed_txn.rawTransaction)

print("Deploying Token Contract... TX:", tx_hash.hex())
```

## 🔹 Step 9: Get Contract Address

```python
tx_receipt = w3.eth.wait_for_transaction_receipt(tx_hash)
contract_address = tx_receipt.contractAddress

print("Contract Address:", contract_address)
```

## 🔹 Step 10: Interact with Contract

```python
contract = w3.eth.contract(address=contract_address, abi=abi)

# Check initial balance
balance = contract.functions.getBalance(account).call()
print("Owner Balance:", balance)
```

## 🔹 Step 11: Transfer Tokens

```python
receiver = "RECEIVER_ADDRESS"

nonce = w3.eth.get_transaction_count(account)

txn = contract.functions.transfer(receiver, 100).build_transaction({
    'from': account,
    'nonce': nonce,
    'gas': 200000,
    'gasPrice': w3.to_wei('10', 'gwei')
})

signed_txn = w3.eth.account.sign_transaction(txn, private_key)
tx_hash = w3.eth.send_raw_transaction(signed_txn.rawTransaction)

print("Transfer TX:", tx_hash.hex())
```

## 🔹 Step 12: Check Updated Balances

```python
owner_balance = contract.functions.getBalance(account).call()
receiver_balance = contract.functions.getBalance(receiver).call()

print("Owner Balance:", owner_balance)
print("Receiver Balance:", receiver_balance)
```
