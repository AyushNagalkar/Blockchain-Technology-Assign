🔹 Step 2: Import Libraries

```python
from web3 import Web3
from solcx import compile_source, install_solc
```

🔹 Step 3: Install Solidity Compiler

```python
install_solc('0.8.0')
```

🔹 Step 4: Connect to Ethereum Network

```python
infura_url = "https://sepolia.infura.io/v3/YOUR_PROJECT_ID"
w3 = Web3(Web3.HTTPProvider(infura_url))

print("Connected:", w3.is_connected())
```

🔹 Step 5: Wallet Configuration

```python
account = "YOUR_WALLET_ADDRESS"
private_key = "YOUR_PRIVATE_KEY"
```

🔹 Step 6: Write Smart Contract

```python
contract_source_code = '''
pragma solidity ^0.8.0;

contract HelloWorld {
    string public message;

    constructor() {
        message = "Hello, Blockchain!";
    }

    function setMessage(string memory newMessage) public {
        message = newMessage;
    }

    function getMessage() public view returns (string memory) {
        return message;
    }
}
'''
```

🔹 Step 7: Compile Contract

```python
compiled_sol = compile_source(contract_source_code)
contract_id, contract_interface = compiled_sol.popitem()

bytecode = contract_interface['bin']
abi = contract_interface['abi']
```

🔹 Step 8: Deploy Contract

```python
HelloWorld = w3.eth.contract(abi=abi, bytecode=bytecode)

nonce = w3.eth.get_transaction_count(account)

transaction = HelloWorld.constructor().build_transaction({
    'from': account,
    'nonce': nonce,
    'gas': 2000000,
    'gasPrice': w3.to_wei('10', 'gwei')
})

signed_txn = w3.eth.account.sign_transaction(transaction, private_key)
tx_hash = w3.eth.send_raw_transaction(signed_txn.rawTransaction)

print("Deploying... TX Hash:", tx_hash.hex())
```

🔹 Step 9: Get Contract Address

```python
tx_receipt = w3.eth.wait_for_transaction_receipt(tx_hash)
contract_address = tx_receipt.contractAddress

print("Contract deployed at:", contract_address)
```

🔹 Step 10: Interact with Contract

```python
contract = w3.eth.contract(address=contract_address, abi=abi)

print("Message:", contract.functions.getMessage().call())
```

🔹 Step 11: Update Contract State

```python
nonce = w3.eth.get_transaction_count(account)

txn = contract.functions.setMessage("Hello from Python").build_transaction({
    'from': account,
    'nonce': nonce,
    'gas': 200000,
    'gasPrice': w3.to_wei('10', 'gwei')
})

signed_txn = w3.eth.account.sign_transaction(txn, private_key)
tx_hash = w3.eth.send_raw_transaction(signed_txn.rawTransaction)

print("Updating message... TX:", tx_hash.hex())
```
