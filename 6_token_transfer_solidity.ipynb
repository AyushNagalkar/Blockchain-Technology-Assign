```python
solidity_code = '''\n
// SPDX-License-Identifier: MIT\n
pragma solidity ^0.8.0;\n
\n
contract SimpleToken {\n
    mapping(address => uint256) public balances;\n
\n
    constructor(uint256 initialSupply) {\n
        balances[msg.sender] = initialSupply;\n
    }\n
\n
    function transfer(address to, uint256 amount) public {\n
        require(balances[msg.sender] >= amount, "Insufficient balance");\n
        balances[msg.sender] -= amount;\n
        balances[to] += amount;\n
    }\n
}\n
'''
```

```python

with open("SimpleToken.sol", "w") as f:
    f.write(solidity_code)

```

```python
print("Solidity contract written to SimpleToken.sol")
```

```python
class SimpleTokenSimulation:
    def __init__(self, owner, initial_supply):
        self.balances = {owner: initial_supply}

    def transfer(self, sender, receiver, amount):
        if self.balances.get(sender, 0) < amount:
            raise ValueError("Insufficient balance")
        self.balances[sender] -= amount
        self.balances[receiver] = self.balances.get(receiver, 0) + amount
```

```python
owner = "Alice"
user1 = "Bob"
user2 = "Charlie"
```

```python
token = SimpleTokenSimulation(owner, 1000)
print("Initial balances:", token.balances)
```

```python
token.transfer(owner, user1, 200)
token.transfer(user1, user2, 50)
```

```python

print("Balances after transfers:", token.balances)
```
